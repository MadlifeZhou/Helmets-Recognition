```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import ImageEnhance
from PIL import Image
import math
```

```python
path='E:\Zhou_Ubuntu\kind2.jpg'
```

```python
raw_img = cv2.imread(path)
raw_img = raw_img[:,:,[2,1,0]]
plt.imshow(raw_img)
print(type(raw_img))
```

```python
image_copy = Image.open(path)
enh_con = ImageEnhance.Contrast(image_copy)
contrast = 2
image_contrasted = enh_con.enhance(contrast)
image_contrasted = np.array(image_contrasted)
plt.imshow(image_contrasted)
print(type(image_contrasted))
```

```python
gray_img = cv2.cvtColor(image_contrasted, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img,'gray')
```

```python
circles = cv2.HoughCircles(gray_img,cv2.HOUGH_GRADIENT,1,20,param1=20,param2=30)
circles = np.uint16(np.around(circles))
print(circles)
for i in circles[0,:]:
    # draw the outer circle    
    c = cv2.circle(image_contrasted,(i[0],i[1]),i[2],(0,0,0),2)
    # draw the center of the circle
# plt.imshow(c)
cv2.imshow('img',c)
cv2.waitKeys(0)
```

```python
c_gray = cv2.cvtColor(c,cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(c_gray,0,255,cv2.THRESH_OTSU)
# contours,hierarchy = cv2.findContours(binary,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = cv2.findContours(binary,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))
plt.imshow(binary,'gray')
```

```python
for i in contours:
    print(cv2.contourArea(i))
```
